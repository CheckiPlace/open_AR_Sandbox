```python
def s_topo(s_frame):
    #a_topo = s_frame[20:200, 20:250]
    a_topo = s_frame.max() - s_frame
    aa_topo = 1.0/np.sum(a_topo) * a_topo
    
    plt.pcolormesh(aa_topo)
    plt.plot(50,160,'ro')
    plt.colorbar()
    return aa_topo
    
```
