```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# set up resolution as given by documentation
res = np.array([1080, 1920, 4])
res
```

```python
# load data and reshape it according to resolution
dat = np.load('dat.npy').reshape([res[0], res[1], res[2]])
```

```python
# extract single channels
r = dat[:,:,2]
g = dat[:,:,1]
b = dat[:,:,0]
a = dat[:,:,3]
```

```python
# alpha channel
a
```

```python
# rearrange single channels for true color image, neglecting unused alpha channel
img = np.stack((r,g,b), axis=2)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(22,14))
ax = ax.ravel()
ax[0].imshow(r, cmap='Reds_r')
ax[0].title.set_text('Red channel')
ax[1].imshow(g, cmap='Greens_r')
ax[1].title.set_text('Green channel')
ax[2].imshow(b, cmap='Blues_r')
ax[2].title.set_text('Blue channel')
ax[3].imshow(img)
ax[3].title.set_text('True color')

plt.show()
```

```python
# save image as png and jpeg
plt.imsave('dat.png', img)
plt.imsave('dat.jpg', img, format='jpg')
```
